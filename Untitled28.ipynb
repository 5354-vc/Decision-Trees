```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
df = pd.read_csv(r'C:\Users\vansh\Downloads\Social_Network_Ads.csv')
df.head()
```

```python
df['Gender'] = df['Gender'].map({'Male':1,'Female':0})
df.head()
```

```python
X= df.drop(columns=['User ID','Purchased'])
y = df['Purchased']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
#Scaling the Data
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train_transform = ss.fit_transform(X_train)
X_test_transform = ss.transform(X_test)
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
tree.fit(X_train_transform,y_train)
y_predict=tree.predict(X_test_transform)
from sklearn.metrics import accuracy_score,f1_score
accuracy_score(y_test,y_predict)
f1_score(y_test,y_predict)
from sklearn.tree import plot_tree
plt.figure(figsize=(30,25))
plot_tree(tree)
plt.show()
```
